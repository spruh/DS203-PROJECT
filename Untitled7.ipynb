# Polynomial Regression
## Temp in india basic regression to see if regression works well for temp rediction
### https://www.kaggle.com/datasets/venky73/temperatures-of-india

```python
dft=pd.read_csv('tempdata.csv')
display(dft)
```

```python
#we know apply a regression model on the month of may india
dft['avg']=dft.iloc[:,2:14].sum(axis=1)/12
display(dft)
```

```python
#after shifting origin
x=dft['YEAR']
x=x.to_numpy()

display(x)
y=dft['avg']
```

```python
from sklearn.preprocessing import PolynomialFeatures

import matplotlib.pyplot as plt
poly = PolynomialFeatures(degree=3, include_bias=False)

poly_features = poly.fit_transform(x.reshape(-1,1))

from sklearn.linear_model import LinearRegression
poly_reg_model = LinearRegression()
poly_reg_model.fit(poly_features, y)
y_predicted = poly_reg_model.predict(poly_features)

plt.figure(figsize=(10, 6))
plt.title('Polynomial Regression', size=16)
plt.scatter(x,y)
plt.plot(x, y_predicted, c='red')
plt.show()



```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y,y_predicted))
print(rmse)
```

#### conclusion

The polynomial of degree three has least rmse and hence. a third degeree polynomial seems a good fit
We see that polynomial  regression isnt perhaps th best way to predict temperature,for instance the latest years are bringing considerable anomalies


# Random Forest
### https://www.kaggle.com/datasets/sudalairajkumar/daily-temperature-of-major-cities
### analysis specific to major cities in india

```python
df2=pd.read_csv('city_temperature.csv')
display(df2)
```

```python
df2=df2.loc[df2['Country']=='India']
display(df2)
df3=df2.iloc[:,3:8]
display(df3)

```

```python
features = pd.get_dummies(df3)
display(features)
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(features['AvgTemperature'])
# Remove the labels from the features
# axis 1 refers to the columns
features= features.drop('AvgTemperature', axis = 1)
# Saving feature names for later use
feature_list = list(features.columns)
# Convert to numpy array
features = np.array(features)
```

```python
display(features)
```

```python

from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
display(test_features)
display(test_labels)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)

```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators =300 , random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python

mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 3), '%.')
```

Now that we have seen excellent accuracy lets try to predict todays temperature


```python
tst_ftr=np.array([[11,25,2022,1,0,0,0]])
np.transpose(tst_ftr)
display(tst_ftr.shape)
predictions = rf.predict(tst_ftr)
errors = abs(predictions - 80.6)
mape = 100 * (errors / 80.6)
accuracy = 100 - mape
print(accuracy)

#NOTE TEMP IN FAREINHIT
#80.6 was yesterdays avg temperature
```

## Classification based models
### https://www.kaggle.com/datasets/mahirkukreja/delhi-weather-data


```python
df4=pd.read_csv('testset.csv')
display(df4)
# drop all rows with any NaN and NaT values
df4=df4.iloc[:,[2,3,6,9]]
df4=df4.dropna()
display(df4)
df6=df4.iloc[:,3]
df5=df4.iloc[:,[0,1,2]]

display(df5)
```

# Support Vector Machine

```python
# Will it rain or not
features=np.array(df5)
labels=np.array(df6)
display(labels)
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 0)
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python

from sklearn import svm


clf = svm.SVC(kernel='linear') 


clf.fit(train_features, train_labels)

y_pred = clf.predict(test_features)
```

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(test_labels, y_pred))
```

# Naive Bayes

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
gnb = GaussianNB()

#Train the model using the training sets
gnb.fit(train_features, train_labels)

#Predict the response for test dataset
y_pred = gnb.predict(test_features)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(test_labels, y_pred))
```
